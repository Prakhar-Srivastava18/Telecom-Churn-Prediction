```python
# CELL 1: Importing the tools that we need
# Pandas is our main data manipulation tool
import pandas as pd
# Numpy helps us do maths and handle missing 'blank' values
import numpy as np
```

```python
# CELL 2: Loading the dataset
# This reads the CSV file from your computer into a Pandas DataFrame called 'df'
df = pd.read_csv('telco_data.csv')
# .head() shows the first 5 rows of our data so we can see what it looks like
print("The first 5 rows of our messy data:")
display(df.head())
```

```python
# CELL 3:Finding the missing values!
# .info() gives us the summary of all columns and tells us if any data types look wrong
print("\nData Info ")
df.info()
 
# If we look at the output of df.info(), the 'TotalCharges' column is listed as an "object" (text).
# But total charges should be numbers. This happens because some cells have empty spaces instead of numbers.
# So, forcing it to be a number. 'coerce' means "if you find a space, turn it into a NaN (Not a Number)"
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Counting exactly how many missing values we have in every column
print("\nMissing Values Count ")
print(df.isnull().sum())
```

```python
# CELL 4: Cleaning the Missing Values & Duplicates
# Since only 11 rows are missing in 'TotalCharges' 
# the safest and easiest thing to do is just drop those 11 rows.
df = df.dropna()

# Now, removing any accidental duplicate rows
df = df.drop_duplicates()

print("\nMissing Values AFTER Cleaning ")
print(df.isnull().sum())
print(f"\nFinal dataset size: {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# CELL 5: Fixing Formatting & Useless Columns
# 'customerID' is unique for everyone and doesn't help us find patterns, so we drop it.
df = df.drop(columns=['customerID'], errors='ignore')

# fixing inconsistent formatting by removing any accidental invisible spaces at the start/end of text
# We loop through all the columns that contain text (objects) and strip the spaces
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.strip()

print("\nAfter dropping customerID and cleaning text ")
display(df.head(5))
```

```python
# CELL 6: Saving the Cleaned Dataset!
# We save our cleaned data to a new file so we don't ruin the original messy file.
# index=False stops Pandas from adding a useless column of row numbers.
df.to_csv('cleaned_telco_data.csv', index=False)

print("\nCleaned dataset saved as 'cleaned_telco_data.csv' ")
```
