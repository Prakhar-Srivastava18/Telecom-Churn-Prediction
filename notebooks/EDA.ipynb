```python
# CELL 1: Import Visualization Tools
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
# Loading the CLEANED dataset
df = pd.read_csv('cleaned_telco_data.csv')
print("Cleaned data loaded")
```

```python
# CELL 2: Visualization 1 - The Target Variable
# Question: How many customers are actually leaving (Churn = Yes)?
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Churn', hue='Churn', palette='Set2', legend=False)
plt.title('Customer Churn Count')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# CELL 3: Visualization 2 - Churn by Contract Type
# Question: Does the type of contract affect if people leave?

plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Contract', hue='Churn', palette='pastel')
plt.title('Churn by Contract Type')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# CELL 4: Visualization 3 - Monthly Charges vs Churn
# Question: Do people who pay more monthly churn more?

plt.figure(figsize=(7, 5))
# We use a boxplot to see the average and look for any weird outliers
sns.boxplot(data=df, x='Churn', y='MonthlyCharges', hue='Churn', palette='Set3', legend=False)
plt.title('Monthly Charges vs Churn (Boxplot)')
plt.show()
```

```python
# CELL 5: Visualization 4 - Tenure (How long they stay)
# Question: At what point in their journey do customers usually leave?

plt.figure(figsize=(8, 5))
# A histogram groups months together into "bins" so we can see trends over time
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', bins=35, palette='viridis')
plt.title('Customer Tenure Distribution by Churn')
plt.xlabel('Tenure (Months)')
plt.show()
```

```python
# CELL 6: Visualization 5 - Correlation Heatmap
# Question: How strongly our numeric variables connected to each other?
# First, we must convert our text 'Churn' (Yes/No) into numbers (1/0)
df['Churn_Numeric'] = df['Churn'].map({'Yes': 1, 'No': 0})

# Select only the columns with numbers to compare
numeric_columns = ['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn_Numeric']
correlation_matrix = df[numeric_columns].corr()

plt.figure(figsize=(8, 6))
# annot=True puts the actual numbers inside the colored boxes
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
